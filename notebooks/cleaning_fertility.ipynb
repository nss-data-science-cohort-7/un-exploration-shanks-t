```python
import pandas as pd
import pandas_flavor as pf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import janitor
from janitor import remove_columns, rename_column, remove_empty, filter_on, groupby_agg, clean_names
```

```python
fertility_rate_df = (
    pd.read_excel('../data/fertility_rates.xls', skiprows=3)
    .clean_names()
    .drop(0)
    .rename_column('period', 'year')
)
```

```python
columns_to_keep = ['country', 'total_fertility', 'mean_age_at_childbearing_', 'year']
fertility_rate_df = fertility_rate_df[columns_to_keep]

fertility_rate_df = fertility_rate_df.query("year != '..' and total_fertility != '..' and mean_age_at_childbearing_ != '..'")
fertility_rate_df.dropna(subset=['year'], inplace=True)
fertility_rate_df.tail(10)
```

```python
# fertility_rate_df[fertility_rate_df['period'].isna()]
```

```python
latest_year = '2020'
fertility_rate_df['year'] = fertility_rate_df['year'].replace('Latest', latest_year)

fertility_rate_df['year'] = fertility_rate_df['year'].str.extract('(\d+)')

fertility_rate_df = fertility_rate_df.astype({'total_fertility': 'float', 'mean_age_at_childbearing_': 'float', 'year': 'int'})
fertility_rate_df = fertility_rate_df.dropna()
fertility_rate_df.head(20)
```

```python
fertility_rate_df.to_csv('../data/fertility_rates_cl.csv', index=False)
fertility_rate_df
fertility_rate_df.columns
```
